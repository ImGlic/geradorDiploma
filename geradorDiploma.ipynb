```
from PIL import Image, ImageDraw, ImageFont
import ipywidgets as widgets
from IPython.display import display
from google.colab import files
import zipfile
import os

participantes = [
    ["Paulo Ricardo Mesquita Rosa Santos", "Tutor Inteligente", "01/02/2024", "30/06/2024", 60],
    ["Vinicius Larsen Santos", "FuzzySec", "01/02/2024", "30/06/2024", 60],
    ["Matheus Oliveira Teixeira", "Tutor Inteligente", "01/02/2024", "30/06/2024", 60],
    ["Eduarda Silva Oliveira", "V-AI", "01/02/2024", "30/06/2024", 60],
    ["Gabriel dos Santos Raimundo", "", "01/02/2024", "30/06/2024", 60],
    ["Paulo Cesar Marinho Dos Santos", "Nibby", "01/02/2024", "30/06/2024", 60],
    ["Isabela dos Santos Estabile", "FuzzySec", "01/02/2024", "30/06/2024", 60],
    ["Caio Henrique Dantas Milão Do Nascimento", "Tutor Inteligente", "01/02/2024", "30/06/2024", 60],
    ["Daniel Alves da Cunha", "See For Me", "01/02/2024", "30/06/2024", 60],
    ["Júlia Victoria Carvalhosa de Amorim", "Nibby", "01/02/2024", "30/06/2024", 60],
    ["Mariana Muniz da Rocha", "V-AI", "01/02/2024", "30/06/2024", 60],
    ["Alessandra Drummond Escocard Morisson", "V-AI", "01/02/2024", "30/06/2024", 60],
    ["Arthur Alen Amorelli Pereira", "V-AI", "01/02/2024", "30/06/2024", 60],
    ["Bernardo Pacheco de Melo", "V-AI", "01/02/2024", "30/06/2024", 60],
    ["Arthur Juan Lima Pinto", "FuzzySec", "01/02/2024", "30/06/2024", 60],
    ["Bernardo Gonçalves Spinelli Martins", "Nibby", "01/02/2024", "30/06/2024", 60],
    ["Jean Marcos Rebello Marinho", "Tutor Inteligente", "01/02/2024", "30/06/2024", 60],
    ["Jefferson Correia de Moura Mello", "Nibby", "01/02/2024", "30/06/2024", 60],
    ["Luiz Marcelo Itapicurú Pereira Costa", "Tutor Inteligente", "01/02/2024", "30/06/2024", 60],
    ["Jhonatan de Sant' Anna Brum", "Nibby", "01/02/2024", "30/06/2024", 60],
    ["José Victor Ferreira da Silva", "FuzzySite", "01/02/2024", "30/06/2024", 60],
    ["Victor José Glicério", "FuzzySite", "01/02/2024", "30/06/2024", 60],
    ["Paulo Elias de Oliveira Chatack", "Tutor Inteligente", "01/02/2024", "30/06/2024", 60]
]

def gerar_certificado(nome, projeto, data_inicio, data_fim, horas):
    certificado = Image.open(template_certificado)
    draw = ImageDraw.Draw(certificado)
    largura_imagem, altura_imagem = certificado.size

    fonte_nome = ImageFont.truetype("Roboto-Bold.ttf", 70)
    fonte_padrao = ImageFont.truetype("Roboto-Bold.ttf", 40)

    habilitacao = f"Declaro que {nome.upper()} participou do {projeto} como membro"
    periodo = f"no período de {data_inicio} a {data_fim}."
    habilitacao2 = f"obtendo o total de duração de {horas} horas, demonstrando dedicação e "
    habilitacao3 = "empenho nos projetos em que foi alocado."
    data = "Emitido em 08/08/2024"

    # Calculando as posições X para centralizar o texto
    x_habilitacao = (largura_imagem - draw.textsize(habilitacao, font=fonte_padrao)[0]) / 2
    x_periodo = (largura_imagem - draw.textsize(periodo, font=fonte_padrao)[0]) / 2
    x_habilitacao2 = (largura_imagem - draw.textsize(habilitacao2, font=fonte_padrao)[0]) / 2
    x_habilitacao3 = (largura_imagem - draw.textsize(habilitacao3, font=fonte_padrao)[0]) / 2
    x_data = (largura_imagem - draw.textsize(data, font=fonte_padrao)[0]) / 2

    y_habilitacao = 900
    y_periodo = 950
    y_habilitacao2 = 1000
    y_habilitacao3 = 1040
    y_data = 1300

    draw.text((x_habilitacao, y_habilitacao), habilitacao, font=fonte_padrao, fill="black")
    draw.text((x_periodo, y_periodo), periodo, font=fonte_padrao, fill="black")
    draw.text((x_habilitacao2, y_habilitacao2), habilitacao2, font=fonte_padrao, fill="black")
    draw.text((x_habilitacao3, y_habilitacao3), habilitacao3, font=fonte_padrao, fill="black")
    draw.text((x_data, y_data), data, font=fonte_padrao, fill="black")

    nome_arquivo = f"certificado_{nome}.png"
    certificado.save(nome_arquivo)
    return nome_arquivo

def on_add_button_click(b):
    nome = nome_input.value
    projeto = projeto_input.value
    data_inicio = data_inicio_input.value
    data_fim = data_fim_input.value
    horas = horas_input.value
    if nome and projeto and data_inicio and data_fim and horas:
        participantes.append((nome, projeto, data_inicio, data_fim, horas))
        nome_input.value = ''
        projeto_input.value = ''
        data_inicio_input.value = ''
        data_fim_input.value = ''
        horas_input.value = ''
    else:
        print("Por favor, preencha todos os campos.")

def on_generate_button_click(b):
    arquivos_certificados = []
    for nome, projeto, data_inicio, data_fim, horas in participantes:
        arquivo = gerar_certificado(nome, projeto, data_inicio, data_fim, horas)
        arquivos_certificados.append(arquivo)

    # Compactando os arquivos em um ZIP
    zip_filename = "certificados.zip"
    with zipfile.ZipFile(zip_filename, 'w') as zipf:
        for arquivo in arquivos_certificados:
            zipf.write(arquivo)

    # Baixando o arquivo ZIP
    files.download(zip_filename)

    # Limpando os arquivos gerados
    for arquivo in arquivos_certificados:
        os.remove(arquivo)
    participantes.clear()

uploaded = files.upload()
for filename in uploaded.keys():
    template_certificado = filename

nome_input = widgets.Text(placeholder='Digite o nome')
projeto_input = widgets.Text(placeholder='Digite o projeto')
data_inicio_input = widgets.Text(placeholder='Digite a data de início (dd/mm/aaaa)')
data_fim_input = widgets.Text(placeholder='Digite a data de fim (dd/mm/aaaa)')
horas_input = widgets.Text(placeholder='Digite o total de horas')
add_button = widgets.Button(description="Adicionar")
generate_button = widgets.Button(description="Gerar Certificados")

add_button.on_click(on_add_button_click)
generate_button.on_click(on_generate_button_click)

display(widgets.VBox([nome_input, projeto_input, data_inicio_input, data_fim_input, horas_input, add_button, generate_button]))
```
